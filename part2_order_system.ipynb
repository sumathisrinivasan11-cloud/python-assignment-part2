```python
# ---- Task 1: Explore the Menu ----

# Step 1: Get all unique categories
categories = []
for details in menu.values():
    if details["category"] not in categories:
        categories.append(details["category"])

# Step 2: Print menu grouped by category
for category in categories:
    print(f"\n===== {category} =====")
    for item, details in menu.items():
        if details["category"] == category:
            status = "[Available]" if details["available"] else "[Unavailable]"
            print(f"{item:<15} ₹{details['price']:.2f}   {status}")

# ---- Statistics ----

# Total number of items
total_items = len(menu)
print("\nTotal menu items:", total_items)

# Total available items
available_items = 0
for details in menu.values():
    if details["available"]:
        available_items += 1
print("Available items:", available_items)

# Most expensive item
most_expensive_item = None
highest_price = 0

for item, details in menu.items():
    if details["price"] > highest_price:
        highest_price = details["price"]
        most_expensive_item = item

print(f"Most expensive item: {most_expensive_item} (₹{highest_price:.2f})")

# Items under ₹150
print("\nItems priced under ₹150:")
for item, details in menu.items():
    if details["price"] < 150:
        print(f"- {item}: ₹{details['price']:.2f}")
```

```python
cart = []
```

```python
def add_to_cart(item_name, quantity):
    # Check if item exists in menu
    if item_name not in menu:
        print(f"Cannot add '{item_name}': Item does not exist in menu.")
        return

    # Check if item is available
    if not menu[item_name]["available"]:
        print(f" Cannot add '{item_name}': Item is unavailable.")
        return

    # Check if item already in cart → update quantity
    for entry in cart:
        if entry["item"] == item_name:
            entry["quantity"] += quantity
            print(f" Updated quantity of {item_name} to {entry['quantity']}")
            return

    # Otherwise add new entry
    cart.append({
        "item": item_name,
        "quantity": quantity,
        "price": menu[item_name]["price"]
    })
    print(f" Added {item_name} × {quantity} to cart.")
```

```python
def remove_from_cart(item_name):
    for entry in cart:
        if entry["item"] == item_name:
            cart.remove(entry)
            print(f"Removed {item_name} from cart.")
            return

    print(f" Cannot remove '{item_name}': Item not in cart.")
```

```python
def update_quantity(item_name, new_quantity):
    for entry in cart:
        if entry["item"] == item_name:
            entry["quantity"] = new_quantity
            print(f"Updated {item_name} quantity to {new_quantity}")
            return

    print(f"Cannot update '{item_name}': Item not in cart.")
```

```python
cart = [
    {"item": "Paneer Tikka", "quantity": 3, "price": 180.0}
]
```

```python
print("\n--- Deducting Stock Based on Cart ---")

for entry in cart:
    item = entry["item"]
    qty_needed = entry["quantity"]
    current_stock = inventory[item]["stock"]

    if current_stock >= qty_needed:
        inventory[item]["stock"] -= qty_needed
        print(f"Deducted {qty_needed} from {item}. New stock: {inventory[item]['stock']}")
    else:
        print(f"Warning: Only {current_stock} units of {item} available. Deducting all.")
        inventory[item]["stock"] = 0
```

```python
print("\n--- Reorder Alerts ---")

for item, details in inventory.items():
    if details["stock"] <= details["reorder_level"]:
        print(f"Reorder Alert: {item} — Only {details['stock']} unit(s) left (reorder level: {details['reorder_level']})")
```

```python
print("\nFinal Inventory (After Deduction):")
print(inventory)

print("\nInventory Backup (Original):")
print(inventory_backup)
```

```python
print("=== Total Revenue Per Day ===")

daily_revenue = {}   # store totals here

for date, orders in sales_log.items():
    total = 0
    for order in orders:
        total += order["total"]
    daily_revenue[date] = total
    print(f"{date}: ₹{total:.2f}")
```

```python
best_day = max(daily_revenue, key=daily_revenue.get)
print(f"\nBest-selling day: {best_day} (₹{daily_revenue[best_day]:.2f})")
```

```python
item_frequency = {}

for orders in sales_log.values():
    for order in orders:
        for item in order["items"]:
            if item not in item_frequency:
                item_frequency[item] = 0
            item_frequency[item] += 1

most_ordered_item = max(item_frequency, key=item_frequency.get)

print(f"\nMost ordered item: {most_ordered_item} ({item_frequency[most_ordered_item]} orders)")
```
